```python
# Move data from dataset into format expected by yolo5
import shutil, os
trainingFiles = os.listdir("/content/YOLO_Realistic_USA_RoadSigns_160classes/train")
validFiles = os.listdir("/content/YOLO_Realistic_USA_RoadSigns_160classes/valid")

for file in trainingFiles:
  if file == "classes.txt":
    pass # Do nothing, classes.txt is useless
  elif file.endswith(".txt"):
    # Get associated jpg file
    jpgFile = file.replace(".txt", ".jpg")
    # Move files into right location with T in front
    imagePath = "/content/dataset/images/train/" + jpgFile
    labelPath = "/content/dataset/labels/train/" + file
    originImage = "/content/YOLO_Realistic_USA_RoadSigns_160classes/train/"+jpgFile
    originLabel = "/content/YOLO_Realistic_USA_RoadSigns_160classes/train/"+file
    shutil.move(originLabel, labelPath)
    shutil.move(originImage, imagePath)


for file in validFiles:
  if file == "classes.txt":
    pass # Do nothing, classes.txt is useless
  elif file.endswith(".txt"):
    # Get associated jpg file
    jpgFile = file.replace(".txt", ".jpg")
    # Move files into right location with T in front
    imagePath = "/content/dataset/images/valid/" + jpgFile
    labelPath = "/content/dataset/labels/valid/" + file
    originImage = "/content/YOLO_Realistic_USA_RoadSigns_160classes/valid/"+jpgFile
    originLabel = "/content/YOLO_Realistic_USA_RoadSigns_160classes/valid/"+file
    shutil.move(originLabel, labelPath)
    shutil.move(originImage, imagePath)
```
